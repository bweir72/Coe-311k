```python
import random as rnd

def appxPi(numSamples):
    counter = 0
    for i in range (0,numSamples,20):
        x = rnd.uniform(-1,1)
        y = rnd.uniform(-1,1)
        if( x**2 + y**2 < 1):
            counter += 1
    
    pi = 4 * counter / numSamples
    return pi
        
```

```python
numSamples = 1000
pi_est = appxPi(numSamples)
print(pi_est)
```
