```python
import numpy as np
baseTen = 2025
quotient = int(baseTen / 2)
remainder = baseTen % 2
binary = np.array([])
binary = np.append(binary, remainder)


while(quotient > 0):
    baseTen = quotient
    quotient = int(quotient / 2)
    remainder = baseTen % 2
    binary = np.insert(binary, 0, remainder)
    
print(binary)

# I used ai to find np.insert
# Is there a key word that makes a new appended number go to the beginning in python
# chatgpt
```
