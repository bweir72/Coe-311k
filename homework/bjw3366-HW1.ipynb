```python
import random as rng
import numpy as np
import matplotlib.pyplot as plt
```

```python
def piError(numSamples):  #I need to generate 10,30,50.. random points per loop and divide the correct amount of samples
    error = np.array([])
    
    for i in range (10,numSamples,20):
        counter = 0
        
        #generates i amount of samples
        for j in range(i): 
            x = rng.uniform(-1,1)
            y = rng.uniform(-1,1)
            
            #verifies it lies inside the circles
            if( x**2 + y**2 < 1):
                counter += 1
        
        pi_est = 4 * counter / i
        error = np.append(error,abs(np.pi - pi_est))
    
    return error
        
```

```python
def main():
    numSamples = 10000
    error = piError(numSamples)
    x = np.arange(10, numSamples, 20)

    plt.figure()
    plt.plot(x, error)
    plt.xlabel('Number of Samples')
    plt.ylabel('Error of Pi')
    plt.title('Pi Error')
    plt.show()
    
    return 0

if __name__ == "__main__":
    main()
```
